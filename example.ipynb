### blbl

始终是以百分比设置视频的宽高, 默认为原视频的 `100%`
由于要保持视频大小比率不变,因此,只有 `width` 有效,  没有`height`这个参数.
`blbl` 有两个参数 `width` 和  `align`,`width` 是嵌入页面的百分比(`默认 100%`), 按百分比进行调整, `align` 只有三种: `left(默认), center, right`
